### **One-sample t-test to check whether the average weight of the rat is equal to 25 kg**


Assume confidence interval is 95% or 0.95 so alpha is 1-0.95 = 0.05

**1.** **Hypothesis**
- H0: The average weight of the rats = 25 kg (μ = 25)
- H1: The average weight of the rats ≠ 25 kg (μ ≠ 25)

**2.** **Distribusi** : Student’s t (two tailed test)

#### Data Description

```python
#find mean, minmax,variance
stats.describe(weight)
```

#### Hypothesis Test - Calculating P-Value

```python
stats.ttest_1samp(weight, popmean =25)
```

```python
from scipy.stats import ttest_1samp

#Average value tested
mean_to_test = 25

#Uji one sample t-test
t_statistic, p_value = ttest_1samp(weight, mean_to_test)

#Interpretation of test results
alpha = 0.05
if p_value < alpha:
    print(f'P-value is {p_value:.4f} and its less than {alpha}, so we can deny the HO Hypothesis')
else:
    print(f'P-value is {p_value:.4f} and its grather than {alpha}, so we cannot deny the HO Hypothesis')
```

## **Interpret the results of the one sample t-test**

we know that:
- if P-value >= alpha(0.05) we can't deny the HO hypothesis
- if P-value < alpha(0.05) we can deny the HO hypothesis

İn this case our P-value is 0.0842 and its grather than alpha(0.05), we can't deny the HO hypothesis that the average rat weight is 25 kg. So, it seems likely that the average rat weight is indeed 25 kg.

### Conclusion:



Hypothesis test proves that the mean or average of the rat weight is equal to 25 kg.
